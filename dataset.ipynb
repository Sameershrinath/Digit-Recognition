```python
import numpy as np
import tkinter as tk
from PIL import Image, ImageDraw
import matplotlib.pyplot as plt

# Initialize a 28x28 white image
canvas_size = 280  # 280x280 pixels (scaled for better drawing experience)
img = Image.new("L", (canvas_size, canvas_size), 255)  # 'L' mode for grayscale
draw = ImageDraw.Draw(img)

# Initialize Tkinter window
root = tk.Tk()
root.title("Draw a 28x28 Image")

# Create a Canvas
canvas = tk.Canvas(root, width=canvas_size, height=canvas_size, bg="white")
canvas.pack()

# Global variable to store the array
saved_array = None  

def paint(event):
    """Draw on the canvas"""
    x1, y1 = (event.x - 10), (event.y - 10)
    x2, y2 = (event.x + 10), (event.y + 10)
    canvas.create_oval(x1, y1, x2, y2, fill="black", outline="black")
    draw.ellipse([x1, y1, x2, y2], fill="black")  # Draw on the PIL image

canvas.bind("<B1-Motion>", paint)  # Bind left mouse drag to paint

def save_and_visualize():
    """Convert to 28x28, save as numpy array, and visualize"""
    global saved_array  # Access the global variable

    resized_img = img.resize((28, 28))  # Resize to 28x28
    saved_array = np.array(resized_img)  # Store in global variable

    # Visualize using matshow()
    plt.matshow(saved_array, cmap="gray")
    plt.show()

    return saved_array  # Return the array for further use

# Add a button to save and visualize
button = tk.Button(root, text="Save & Visualize", command=save_and_visualize)
button.pack()

root.mainloop()

# After closing the window, you can use saved_array
if saved_array is not None:
    print("Saved Array Shape:", saved_array.shape)
```
